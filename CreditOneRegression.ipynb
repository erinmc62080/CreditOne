```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn import linear_model
```

```python
#removing what I don't want 
regression_data.drop(['ID','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','EDUCATION_other','EDUCATION_graduate school','EDUCATION_high school','EDUCATION_university'], axis=1, inplace= True)
```

```python
regression_data.head()
```

```python
#choosing my features 
X=regression_data.iloc[:,0:17]
print('Summary of feature sample')
X.head()

```

```python
#dependent variable 
y=regression_data['CREDIT LIMIT_high']
y.head()
```

```python
print(y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
algosClass = []
```

```python
algosClass.append(('Random Forest',RandomForestRegressor()))
```

```python
algosClass.append(('Linear Regression',LinearRegression()))
```

```python
algosClass.append(('Support Vector Regression',SVR()))
```

```python
#regression
results = []
names = []
for name, model in algosClass:
    result = cross_val_score(model, X,y, cv=3, scoring='r2')
    names.append(name)
    results.append(result)
```

```python
for i in range(len(names)):
    print(names[i],results[i].mean())
```
